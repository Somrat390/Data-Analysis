```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
#Plotting a simple function
price = [48000,15000,50000,0] ## this is a numerical data
year = [2016,2017,2018,2019] ## this is a categorical data

##now use plot function to draw 2D plot
plt.plot(year,price)  ## year is x and price is y axis
```

```python
# now plot from a dataset
batsman = pd.read_csv("/content/sharma-kohli.csv")
```

```python
batsman
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# now plotting mutiple plot
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
## Label Tittles
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title("Virat Kohli vs RG Sharma Carrer Comparision")
plt.xlabel("Season")
plt.ylabel("Score")
```

```python
# color(hex) we can use hex code also to color the line
plt.plot(batsman['index'],batsman['V Kohli'], color='green')
plt.plot(batsman['index'],batsman['RG Sharma'], color = "black")

plt.title("Virat Kohli vs RG Sharma Carrer Comparision")
plt.xlabel("Season")
plt.ylabel("Score")
```

```python
# linestyle(style,width)
plt.plot(batsman['index'],batsman['V Kohli'], color='green', linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'], color = "black", linestyle='dashed')

plt.title("Virat Kohli vs RG Sharma Carrer Comparision")
plt.xlabel("Season")
plt.ylabel("Score")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color='green', linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'], color = "black", linestyle='dotted')

plt.title("Virat Kohli vs RG Sharma Carrer Comparision")
plt.xlabel("Season")
plt.ylabel("Score")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color='green', linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'], color = "black", linestyle='dashdot')

plt.title("Virat Kohli vs RG Sharma Carrer Comparision")
plt.xlabel("Season")
plt.ylabel("Score")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color='green', linestyle='dashdot',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'], color = "black", linestyle='dashdot',linewidth=2)

plt.title("Virat Kohli vs RG Sharma Carrer Comparision")
plt.xlabel("Season")
plt.ylabel("Score")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color='green', linestyle='dashdot',linewidth=3, marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'], color = "black", linestyle='dashdot',linewidth=2, marker ='o')

plt.title("Virat Kohli vs RG Sharma Carrer Comparision")
plt.xlabel("Season")
plt.ylabel("Score")
```

```python
# legend it will label the column
plt.plot(batsman['index'],batsman['V Kohli'], color='green', linestyle='dashdot',linewidth=3,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color = "black", linestyle='dashdot',linewidth=2,label='Sharma')

plt.title("Virat Kohli vs RG Sharma Carrer Comparision")
plt.xlabel("Season")
plt.ylabel("Score")

plt.legend(loc = 'best') #it locate at right,left,best,center
```

```python
#to limit the axis beacuse of outlier
price = [48000,15000,50000,450000]
year = [2016,2017,2018,2019]


plt.plot(year,price)
plt.ylim(0,100000)
plt.xlim(2016,2018)
```

```python
#grid
# legend it will label the column
plt.plot(batsman['index'],batsman['V Kohli'], color='green', linestyle='dashdot',linewidth=3,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color = "black", linestyle='dashdot',linewidth=2,label='Sharma')

plt.title("Virat Kohli vs RG Sharma Carrer Comparision")
plt.xlabel("Season")
plt.ylabel("Score")

plt.legend(loc = 'best')
plt.grid()  # show these box in plot
plt.show()  # without it others editor will not show the graph so whave to used it must
```

```python
# size
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
# drawing a scatter plot of total_bill and tips & also show the size of pepole
plt.scatter(tips['total_bill'],tips['tip'], s=tips['size']*20) # its multiply the size with 20
plt.title("Total bill vs tip plot ")
plt.xlabel("total_bill")
plt.ylabel("tip")
```

# Scatter plot without using scatter

# Pie Chart
## Univariate/Bivariate Analysis
## Categorical vs Numerical
## Use Case - To Find Contribution on a standard scale.

```python
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)
plt.show()
```

```python
data = pd.read_csv("/content/gayle-175.csv")
data.head(5)
```

```python
plt.pie(data['batsman_runs'],labels=data['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','red','black'],explode=[0.2,0,0.1,0,0,0])
plt.show()
```
